# EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

df = pd.read_csv(processed_csv_path)

print("\nFirst 5 Rows")
display(df.head())
```

```python

df.info()

display(df.describe())

print(df.isnull().sum())
```

I will now rename num to target

```python
df.rename(columns={'num': 'target'}, inplace=True)

print(df.columns)
```

We have to convert the target variable into binary (just 1 & 0):

```python
df['target'] = df['target'].apply(lambda x: 1 if x > 0 else 0)

target_counts = df['target'].value_counts()
target_percentage = df['target'].value_counts(normalize=True) * 100
```

```python
plt.figure(figsize=(8, 6))

sns.countplot(x='target', data=df, palette='viridis')

plt.title('Distribution of Heart Disease (Binary)', fontsize=16)
plt.xlabel('Heart Disease (0 = Absent, 1 = Present)', fontsize=12)
plt.ylabel('Patient Count', fontsize=12)

plt.show()
```

### I will now refine EDA with new labels 

```python
# Chest Pain
cp_map = {
    1.0: 'Typical Angina',
    2.0: 'Atypical Angina',
    3.0: 'Non-Anginal Pain',
    4.0: 'Asymptomatic'
}
df['cp_label'] = df['cp'].map(cp_map)

# Thalassemia 

thal_map = {
    3.0: 'Normal',
    6.0: 'Fixed Defect',
    7.0: 'Reversible Defect'
}
df['thal_label'] = df['thal'].map(thal_map)

# SLOPE

slope_map = {
    1.0: 'Upsloping',
    2.0: 'Flat',
    3.0: 'Downsloping'
}
df['slope_label'] = df['slope'].map(slope_map)

# Resting ECG

restecg_map = {
    0.0: 'Normal',
    1.0: 'ST-T Wave Abnormality',
    2.0: 'LV Hypertrophy'
}
df['restecg_label'] = df['restecg'].map(restecg_map)

display(df.head())
```

#### I have now created new labels. I will now try to create a few plots through which we can see how categorical features relate to our target variable. 

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 16))
fig.suptitle('Categorical Feature Analysis vs. Heart Disease', fontsize=20, y=1.03)

# Chest Pain (cp) vs Target 
sns.countplot(ax=axes[0, 0], x='cp_label', hue='target', data=df, palette='Greens')
axes[0, 0].set_title('Heart Disease by Chest Pain Type', fontsize=14)
axes[0, 0].set_xlabel('Chest Pain Type')
axes[0, 0].legend(title='Heart Disease', labels=['Absent (0)', 'Present (1)'])

# Thalassemia (thal) vs Target
sns.countplot(ax=axes[0, 1], x='thal_label', hue='target', data=df, palette='Purples')
axes[0, 1].set_title('Heart Disease by Thalassemia', fontsize=14)
axes[0, 1].set_xlabel('Thalassemia')
axes[0, 1].legend(title='Heart Disease', labels=['Absent (0)', 'Present (1)'])

# ST Slope (slope) vs Target
sns.countplot(ax=axes[1, 0], x='slope_label', hue='target', data=df, palette='Reds')
axes[1, 0].set_title('Heart Disease by ST Slope', fontsize=14)
axes[1, 0].set_xlabel('ST Slope')
axes[1, 0].legend(title='Heart Disease', labels=['Absent (0)', 'Present (1)'])

# Resting ECG (restecg) vs Target
sns.countplot(ax=axes[1, 1], x='restecg_label', hue='target', data=df, palette='Blues')
axes[1, 1].set_title('Heart Disease by Resting ECG', fontsize=14)
axes[1, 1].set_xlabel('Resting ECG')
axes[1, 1].legend(title='Heart Disease', labels=['Absent (0)', 'Present (1)'])

plt.tight_layout(rect=[0, 0, 1, 1])
plt.show()
```

### These plots show some insight. They are as follows:

1. Heart Disease by Chest Pain Type: This looks like a very strong predictor. Patients with 'Asymptomatic' pain are far more likely to have heart disease (Present = 1) than not. 

2. Heart Disease by Thalassemia: Patients with 'Reversible Defect' are much more likely to get heart disease. However, people who are 'Normal' are more likely to have no heart disease. 

3. Heart Disease by ST Slope: A 'Flat' slope seems to correlate to a strong presence of heart disease while an 'Upsloping' slope seems to correlate with the absence of heart disease. 

4. Heart Disease by Resting ECG: Out of all four, this seems a little less indecisive in my opinion. 'Normal' and 'LV Hypertrophy' don' seem to have huge differences between their splits.

### Preprocessing

We will now have to preprocess data for data modelling. Few things I will be focusing on are:

1. Seperate feature (x) and target (y)
2. Converting categorical variables into 0/1 dummy columns so that the model can understand them.
3. Train/test split
4. We will try to scale numeric features such as age and chol so that they are on a similar scale which would help logistic regression and KNN produce better results. 

Let's first start with the seperate features 


```python
y = df['target']

X = df.drop(columns=[
    'target',
    'cp_label',
    'thal_label',
    'slope_label',
    'restecg_label'
])  

print("Features")
display(X.head())

print("Target")
display(y.head())
```

We will now convert categorical variables into dummy variables using pandas (one-hot encoding)

```python
categorical_cols = [
    'cp',
    'restecg',
    'slope',
    'thal'
]

for col in categorical_cols:
    X[col] = X[col].astype(str)

X_processed = pd.get_dummies(X, columns=categorical_cols)

print("Processed Features (X)")
display(X_processed.head())

print("\nNew Columns")
print(X_processed.columns)
```

We will drop thal_nan column as it is a missing value

```python
if 'thal_nan' in X_processed.columns:
    X_processed = X_processed.drop(columns=['thal_nan'])
    
print(X_processed.columns)
```

Now, train/test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_processed, 
    y, 
    test_size=0.2, 
    random_state=42
)

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

Finally, we will use a scaler to scale features like age (29-77) and chol (126-564). As they have very different ranges. 

Before, that we will take care of the missing values from ca through imputation (troubleshooting)

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')

imputer.fit(X_train[['ca']])

X_train[['ca']] = imputer.transform(X_train[['ca']])
X_test[['ca']] = imputer.transform(X_test[['ca']])

print("NaNs in X_train after imputation:", X_train.isnull().sum().sum())
print("NaNs in X_test after imputation:", X_test.isnull().sum().sum())
print("Imputation complete.")
```

```python
from sklearn.preprocessing import StandardScaler

numerical_cols = [
    'age',
    'trestbps',
    'chol',
    'thalach',
    'oldpeak',
    'ca'
]

scaler = StandardScaler()

scaler.fit(X_train[numerical_cols])

X_train[numerical_cols] = scaler.transform(X_train[numerical_cols])
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])

print("\nScaled X_train")
display(X_train.head())
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

log_model = LogisticRegression(random_state=42)

log_model.fit(X_train, y_train) 

y_pred_log = log_model.predict(X_test)

print("Logistic Regression Results")

accuracy_log = accuracy_score(y_test, y_pred_log)
print(f"Accuracy: {accuracy_log * 100:.2f}%")

print("\nClassification Report:")
print(classification_report(y_test, y_pred_log))
```

#### Analysis

These are strong results. 

1. The overall accuracy is 83.61%. The model correctly predicted the outcome for ~84% of unseen patients in this test set. 
2. The f1 scores (0.82 for absent & 0.85 for present) are very close. This is good as it shows that the model is good at identifying both sick and healthy patients. 
3. Class 1 (present - sick): Recall is 0.88 (88%): This means that all the patients who actually had the heart disease, this model was able to find 88% of them. This is good. 
4. Class 2 (absent - healthy): The precision is 0.85 (85%): When the model predicted a patient did not have heart disease, it was correct 85% of the time. It is trustworthy to give a healthy prediction. 

Therefore, I think this shows a strong ability and balanced accuracy in identifying patients with heart disease. 

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=5)

knn_model.fit(X_train, y_train)

y_pred_knn = knn_model.predict(X_test)

print("K-Nearest Neighbors (KNN) Results (k=5)")

accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f"Accuracy: {accuracy_knn * 100:.2f}%")

print("\nClassification Report:")
print(classification_report(y_test, y_pred_knn))
```

## Unsupervised Learning

### K-means

```python
from sklearn.cluster import KMeans

X_final = pd.concat([X_train, X_test])

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_final)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
k = 3

kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(X_final)

X_clustered = X_final.copy()
X_clustered['Cluster'] = clusters

print(X_clustered.head())
```

```python
cluster_profiles = X_clustered.groupby('Cluster').mean()

print(cluster_profiles)
```
